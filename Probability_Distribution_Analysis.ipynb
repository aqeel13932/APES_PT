```python
import numpy as np
np.set_printoptions(precision=2, linewidth=200)
```

```python
r,c = 4,4
x = np.zeros((r,c))
x
```

```python
def Get_Surronding(location):
    neighbours=[]
    totalsides=0
    val =0
    if location[0]>=1:
        neighbours.append((location[0]-1,location[1]))
        totalsides+=1
        
    if location[0]<(r-1):
        neighbours.append((location[0]+1,location[1]))
        totalsides+=1
        
    if location[1]>=1:
        neighbours.append((location[0],location[1]-1))
        totalsides+=1
        
    if location[1]<(c-1):
        neighbours.append((location[0],location[1]+1))
        totalsides+=1
        
    totalsides+=1
    return 1.0/totalsides,neighbours
```

```python
G = np.zeros((x.size,x.size))
G
```

```python
def nl(indx):
    return np.ravel_multi_index(indx,x.shape)
```

```python
for loc,__ in np.ndenumerate(x):
    v,locs = Get_Surronding(loc)
    G[nl(loc),nl(loc)] = v
    for l in locs:
        G[nl(loc),nl(l)]=v
```

```python
G = G.dot(G)
```
